```python
class DataPipeLine:
    def fit(self, df):
        self.medians = df[['HouseFloor', 'KitchenSquare']].median()
        self.medians['LifeSquareRate'] = (df['LifeSquare'] / df['Square']).median()

    def transform(df):
        #кол-во комнат и этажность дома в целое
        df[['Rooms', 'HouseFloor']] = df[['Rooms', 'HouseFloor']].astype('int8')
        
        # дополнительный признак, что жилая площадь былла nan
        df['LifeSquare_is_nan'] = 0
        df.loc[df['LifeSquare'].isna(), 'LifeSquare_is_nan'] = 1
        
        
dp = DataPipeLine()
dp.fit(df)
dp.medians
        
        
        
```
